# Sensors Table

```python
import pandas as pd
import numpy as np
```

```python
a = pd.read_csv('./data/prepared/air.csv')
b = pd.read_csv('./data/prepared/noise.csv')
c = pd.read_csv('./data/prepared/water.csv')
d = pd.read_csv('./data/prepared/weather.csv')
```

```python
a['sensor_type'] = 'air'
b['sensor_type'] = 'noise'
c['sensor_type'] = 'water'
d['sensor_type'] = 'weather'
```

```python
air_sensors = a.groupby(['zone', 'sensor_id', 'sensor_type',
                         'vendor', 'latitude', 'longitude']).count().reset_index().iloc[:, :6]
```

```python
noise_sensors = b.groupby(['zone', 'sensor_id', 'sensor_type',
                           'vendor', 'latitude', 'longitude']).count().reset_index().iloc[:, :6]
```

```python
water_sensors = c.groupby(['zone', 'sensor_id', 'sensor_type',
                           'vendor', 'latitude', 'longitude']).count().reset_index().iloc[:, :6]
```

```python
weather_sensors = d.groupby(['zone', 'sensor_id', 'sensor_type',
                             'vendor', 'latitude', 'longitude']).count().reset_index().iloc[:, :6]
```

```python
merge_1 = pd.concat([air_sensors, noise_sensors, water_sensors, weather_sensors],
                    ignore_index=True)
```

```python
merge_1
```

```python
sensor_table = merge_1.groupby(['zone', 'sensor_id', 'sensor_type',
                             'vendor', 'latitude', 'longitude']).count().reset_index()

sensor_table
```

```python
sensor_table.to_csv('./data/prepared/sensor_table.csv', index=False)
```
